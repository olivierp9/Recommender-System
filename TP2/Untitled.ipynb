```r
#Fonctions de cosinus 
cosinus.vm <- function(v,m) { 
    n <- sqrt(colSums(m^2)); 
    (v %*% m)/(n * sqrt(sum(v^2)))
}

# Fonction de la corrélation 
# Nous avons décidé de faire notre propre code au lieu d'utiliser la fonction pré-défini de R afin d'avoir une visibilité
# sur les opérations de la fonction. 
cor.vm <- function(v,m) {
  v_a <-sum(v)/sum(v>0);
  v_i <- colMeans.sparse(m);
  m.center <- t(t(m) - v_i)*(m>0);
  v.center <- (v-v_a)*(v>0);
  n <- colSums(m.center^2);
  (v.center %*% m.center)/sqrt(n * sum(v.center^2))
}


# Trouve les indexes des premieres 'n' valeurs maximales d'une matrice
max.nindex <- function(m, n=5) {
    i <- order(m, decreasing=TRUE)
    return(i[1:n])
}
# Trouve les indexes des premieres 'n' valeurs minimales d'une matrice
min.nindex <- function(m, n=5) {
    i <- order(m)
    return(i[1:n])
}

# col/row mean for sparse matrix. Il est nécessaire de faire ces fonctions puisque la moyenne d'une ligne/colonne d'une matrice
# sparse ne doit que tenir compte des valeur > 0
colMeans.sparse <- function(m) {colSums(m)/colSums(m>0)}
rowMeans.sparse <- function(m) {temp<-rowSums(m)/rowSums(m>0);temp[rowSums(m)==0]<-0;temp}
means.sparse <- function(v) {sum(v)/sum(v>0)}

# Fonction qui calcule l'erruer moyenne absolue et l'erreur quadratique moyen
mae.rmse <- function(m, m.hat) {
    return(list(mae=mean(abs((m - m.hat)[!is.na(m)]), na.rm=T), rmse=sqrt(mean((m - m.hat)^2, na.rm=T))))
}
```
