```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import warnings
warnings.filterwarnings('ignore')

%matplotlib inline
```

```python
dataset = pd.read_csv("u.data.csv",sep='|',names="user_id,item_id,rating,timestamp".split(","))

dataset.user_id = dataset.user_id.astype('category').cat.codes.values
dataset.item_id = dataset.item_id.astype('category').cat.codes.values
dataset = dataset.iloc[1:,:]
```

```python
dataset.columns
```

```python
df_item = pd.read_csv('u.item.csv', sep='|')
dataset = dataset.merge(df_item, left_on='item_id', right_on='movie id ', how='outer')
dataset = dataset.drop([' video release date '], axis=1)
```

```python
dataset.head()
```

```python
len(dataset.user_id.unique()), len(dataset.item_id.unique())
```

```python
dataset = dataset.dropna(axis=0, how='any', thresh=None, subset=None, inplace=False)
```

```python
dataset.isnull().any()
```

```python
from sklearn.model_selection import train_test_split
train, test = train_test_split(dataset, test_size=0.2)
```

```python
train
```

```python
import keras
from IPython.display import SVG
from keras.optimizers import Adam
from keras.utils.vis_utils import model_to_dot
from keras import regularizers

n_users, n_movies = len(dataset.user_id.unique()), len(dataset.item_id.unique())
n_latent_factors = 3
```

```python
movie_input = keras.layers.Input(shape=[1],name='Item')
movie_embedding = keras.layers.Embedding(n_movies + 1, 5, name='Movie-Embedding')(movie_input)
movie_vec = keras.layers.Flatten(name='FlattenMovies')(movie_embedding)

user_input = keras.layers.Input(shape=[1],name='User')
user_vec = keras.layers.Flatten(name='FlattenUsers')(keras.layers.Embedding(n_users + 1, 5,name='User-Embedding')(user_input))

genre_input = keras.layers.Input(shape=[19],name='Genre')
genre_vec = keras.layers.Dense(19,input_shape=(19,), activation='relu',kernel_regularizer=regularizers.l2(0.0),
                activity_regularizer=regularizers.l1(0.0),name='Genre-Dense')(genre_input)


concat = keras.layers.merge([movie_vec, user_vec,genre_vec], mode='concat',name='Concat')
result = keras.layers.Dense(10, activation='relu',name='Activation')(concat)

result2 = keras.layers.Dense(1, activation='relu',name='Activation2')(result)

model = keras.Model([user_input, movie_input, genre_input], result2)

model.compile(optimizer='adam',loss= 'mean_squared_error')
```

```python
history = model.fit([train.user_id, train.item_id, train.iloc[:,-19:]], train.rating, epochs=100)
```

```python
y_hat = model.predict([test.user_id, test.item_id, test.iloc[:,-19:]])
y_true = test.rating
```

```python
from sklearn.metrics import mean_absolute_error
mean_absolute_error(y_true, y_hat)
```
