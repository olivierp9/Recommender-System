```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import warnings
warnings.filterwarnings('ignore')

%matplotlib inline
```

```python
dataset = pd.read_csv("u.data.csv",sep='|',names="user_id,item_id,rating,timestamp".split(","))
dataset=dataset.drop('timestamp',axis=1)
dataset = dataset.iloc[1:,:]

datauser = pd.read_csv("u.user.csv",sep='|')
datauser = datauser.rename(columns={'id ': 'user_id',' age ':'age',' gender ':'gender',' job ':'job'})

df_item = pd.read_csv('u.item.csv', sep='|')
df_item  = df_item.drop([' video release date ',' movie title ', ' release date ',' IMDb URL '], axis=1)
df_item = df_item.rename(columns={'movie id ': 'item_id'})
```

```python
m = np.zeros((943,1682))
for i in range(len(dataset)):
    m[int(dataset.iloc[i,0])-1,int(dataset.iloc[i,1])-1]=int(dataset.iloc[i,2])
```

```python
dataset.item_id = dataset.item_id.astype('int16')
dataset.user_id = dataset.user_id.astype('int16')
dataset.item_id = dataset.item_id.astype('category')
dataset.user_id = dataset.user_id.astype('category')
```

```python
datauser.user_id = datauser.user_id.astype('int16')
datauser.user_id = datauser.user_id.astype('category')
df_item.item_id = df_item.item_id.astype('int16')
df_item.item_id = df_item.item_id.astype('category')
```

```python
#on regarde s'il ny a pas de NA
dataset.isnull().any()
```

```python
from sklearn.model_selection import train_test_split
from sklearn.model_selection import KFold
```

```python
import keras
from IPython.display import SVG
from keras.optimizers import Adam
from keras.utils.vis_utils import model_to_dot
from keras import regularizers

n_users, n_movies = len(dataset.user_id.unique()), len(dataset.item_id.unique())
# on choisie 9facteurs latents
n_latent_factors = 9
```

```python
# on cree le modele

movie_input = keras.layers.Input(shape=[1],name='Item')
movie_embedding = keras.layers.Embedding(n_movies+1, n_latent_factors, name='Movie-Embedding')(movie_input)
movie_vec = keras.layers.Flatten(name='FlattenMovies')(movie_embedding)

user_input = keras.layers.Input(shape=[1],name='User')
user_vec = keras.layers.Flatten(name='FlattenUsers')(keras.layers.Embedding(n_users+1, n_latent_factors,name='User-Embedding')(user_input))


result = keras.layers.merge([movie_vec, user_vec], mode='dot',name='dot')


model = keras.Model([user_input, movie_input], result)

model.compile(optimizer='adam',loss= 'mean_squared_error')
```

```python
from sklearn.metrics import mean_absolute_error
kf = KFold(n_splits=5, random_state=1, shuffle=True)
split5 = kf.split(dataset)
```

```python
import csv
from copy import copy
```

On effectue la 5-fold avec les bons parametres calcules dans le grid search et les donnees sont ecrites dans des fichiers csv

```python
k=1
for train_index, test_index in kf.split(dataset):
    # creation du modele
    keras.backend.clear_session()
    movie_input = keras.layers.Input(shape=[1],name='Item')
    movie_embedding = keras.layers.Embedding(n_movies, n_latent_factors, name='Movie-Embedding')(movie_input)
    movie_vec = keras.layers.Flatten(name='FlattenMovies')(movie_embedding)
    user_input = keras.layers.Input(shape=[1],name='User')
    user_vec = keras.layers.Flatten(name='FlattenUsers')(keras.layers.Embedding(n_users, n_latent_factors,name='User-Embedding')(user_input))
    result = keras.layers.merge([movie_vec, user_vec], mode='dot',name='dot')
    model = keras.Model([user_input, movie_input], result)
    model.compile(optimizer='adam',loss= 'mean_squared_error')

    train, test = dataset.iloc[train_index,:], dataset.iloc[test_index,:]
    history = model.fit([train.user_id, train.item_id], train.rating, epochs=25,batch_size=32)
    y_hat = model.predict([test.user_id, test.item_id])
    y_true = test.rating
    print(mean_absolute_error(y_true, y_hat))
    m_test = copy(m)
    # on impute les donnes
    for i in range(len(test_index)):
        m_test[dataset.iloc[test_index[i],0]-1,dataset.iloc[test_index[i],1]-1]=0
    for i in range(m_test.shape[0]):
        for j in range(m_test.shape[1]):
            if m_test[i,j] == 0:
                m_test[i,j] = model.predict([datauser.user_id.iloc[i:i+1], df_item.item_id.iloc[j:j+1]])[0][0]
    last_movie = df_item.tail(1).item_id
    for i in range(m_test.shape[0]):
        m_test[i,1681]= model.predict([datauser.user_id.iloc[i:i+1], last_movie])[0][0]
    last_user = datauser.tail(1).user_id
    for i in range(m_test.shape[1]):
        m_test[942,i]= model.predict([last_user, df_item.item_id.iloc[i:i+1]])[0][0]
    m_test[942,1681]=model.predict([last_user,last_movie])[0][0]
    
    # on ecrit les donnees dans des csv
    with open("out/embed/m5_"+str(k)+".csv","w+") as my_csv:
        csvWriter = csv.writer(my_csv,delimiter=',')
        csvWriter.writerows(m_test)
    with open("out/embed/train5_"+str(k)+".csv","w+") as my_csv:
        csvWriter = csv.writer(my_csv,delimiter=',')
        csvWriter.writerow(train_index)
    with open("out/embed/test5_"+str(k)+".csv","w+") as my_csv:
        csvWriter = csv.writer(my_csv,delimiter=',')
        csvWriter.writerow(test_index)
    k+=1
```
